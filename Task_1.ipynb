```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python
#Import Dataset
df=pd.read_csv("sales_data.csv")
```

```python
#Print Sample Row
df.sample(5)
```

```python
#Total Rows and Columns in the DataSet
df.shape
```

```python
#Dataset Description
df.info()
```

```python
df.describe()
```

```python
#Missing Values by Column
missing_values_count=df.isnull().sum()
print(missing_values_count)
```

```python
#Total Missing Values in the Dataset
print(missing_values_count.sum())
```

```python
#Fill Missing values with mean
df["Customer_Age"]=df["Customer_Age"].fillna(df["Customer_Age"].mean())
print("Total Null Values in Customer Age Column:",df["Customer_Age"].isnull().sum())
```

```python
#Fill Missing values with Numeric Values like '0'
df['Age_Group']=df['Age_Group'].fillna(0)
print("Total Null Values in Age Group Column:",df["Customer_Age"].isnull().sum())
```

```python
#Fill the Missing values with Backward fill
df['Product']=df['Product'].bfill(axis=0)
print("Total Null Values in Product Column:",df["Product"].isnull().sum())
```

```python
#Fill the Missing values with Forward Fill
df['Order_Quantity']=df['Order_Quantity'].ffill(axis=0)
print("Total Null Values in Order Quantity Column:",df["Order_Quantity"].isnull().sum())
```

```python
df=df.dropna(axis=0)
print("Total Null Values in the Dataset:",df.isnull().sum().sum())
```

```python
print("Total Duplicate Rows in the Dataset:",df.duplicated().sum())
```

```python
#Remove Duplicate Rows in the Dataset
df.drop_duplicates(inplace=True)
print("Total Duplicate Rows in the Dataset:",df.duplicated().sum())
```

```python
#Standardize Text Values
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
categorical=['Month','Customer_Gender','Country','State','Product_Category','Sub_Category','Product']
df[categorical]=df[categorical].apply(label.fit_transform)
df.sample(5)
```

```python
# Convert DATE to datetime format, handling different formats
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Format all valid dates to dd-mm-yyyy
df['Date'] = df['Date'].dt.strftime('%y/%m/%d')

# Print the updated DataFrame
print(df['Date'].head())
```

```python
#Clean and Rename the Columns
df.columns=(df.columns.str.strip().str.lower().str.replace(' ','_'))
df.columns
```

```python
#Check the Datatypes of the each column
df.info()
```

```python
#Fixed the Date datatype
df["date"]=pd.to_datetime(df["date"])
df.info()
```
